```python

#funcations to process and clean data without  a loop
def cleaner1(value):
    # print(value,type(value))
    if value=='?':
        return np.nan
    return float(value)

#apply only changes to the  particular column
df['normalized-losses']=df['normalized-losses'].apply(cleaner1)

```

```python
#to replace all the ? in the data set
df.replace('?',np.nan,inplace=True)
```

```python
#to check the missing data
df.isnull().sum().plot(kind='pie')
```
